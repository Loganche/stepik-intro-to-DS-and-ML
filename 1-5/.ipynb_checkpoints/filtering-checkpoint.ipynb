## Steps

```python
students_performance = students_performance \
            .rename(columns=
                   {'parental level of education' : 'parental_level_of_education',
                    'test preparation course' : 'test_preparation_course',
                    'math score' : 'math_score',
                    'reading score' : 'reading_score',
                    'writing score' : 'writing_score',
                   })
```

```python
students_performance.head()
```

```python
students_performance.query("gender == 'female' & writing_score > 70").head()
```

```python
writing_query = 80
```

```python
students_performance.query("writing_score > @writing_query").head()
```

```python
score_cols = [i for i in list(students_performance) if 'score' in i]
```

```python
students_performance[score_cols].head()
```
