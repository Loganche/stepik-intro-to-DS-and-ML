```python
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
# time series
stock_data = pd.read_csv('https://raw.githubusercontent.com/PacktPublishing/Pandas-Cookbook/master/data/amzn_stock.csv', index_col='Date', parse_dates=True)
stock_data.head()
```

```python
stock_data.index
```

```python
stock_data['2010']
```

```python
stock_data['2010-02':'2010-06']
```

```python
stock_data.resample('12h').asfreq()
```

```python
stock_data.resample('1w').mean() # mean for the week
```

```python
# оконная функция: катящаяся
# мы применяем функцию к N выбранным строкам
# и даже если у нас сместились эти строки,
# то на них как окно продолжает действовать эта функция
stock_data.rolling(3).mean()
```

```python
stock_data.rolling(3, min_periods=1).mean() # минимальное достаточное кол-во значений для вычислния mean
```

```python
# оконная функция: расширяющаяся
# может менять размер
# среднее в строке считается по кол-ву строк сверху
stock_data.expanding(3).mean()
```

```python
# оконная функция: эскпоненциально взвешенная
# может менять размер
# среднее в строке считается по кол-ву строк сверху
# но еще учитывается вес alpha - насколько близко строка к подсчитывемой
stock_data.ewm(alpha=0.7).mean()
```

```python
ns = stock_data['Open'].rolling(10, min_periods=1).mean()
```

```python
# благодаря оконной функции
# график сгладился
stock_data['Open'].plot()
ns.plot()
```
